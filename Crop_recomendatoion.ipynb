```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
crop_data = pd.read_csv("/content/Crop_Recommendation.csv", header = None)
```

```python
crop_data = crop_data.iloc[1:, :]
```

```python
crop_data[7].value_counts()
```

```python
X = crop_data.drop(columns = 7, axis = 1)
Y = crop_data[7]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1, stratify = Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
le = LabelEncoder()
```

```python
le.fit(Y_train)
```

```python
label_mapping = dict(zip(le.classes_, range(len(le.classes_))))
print(label_mapping)
```

```python
Y_train_encoded = le.fit_transform(Y_train)
Y_test_encoded = le.fit_transform(Y_test)
```

```python
model = svm.SVC()
```

```python
model.fit(X_train, Y_train_encoded)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
mse = mean_squared_error(Y_train_encoded, X_train_prediction)
print('Mean Squared Error:', mse)
```

```python
r2 = r2_score(Y_train_encoded, X_train_prediction)
print('R-squared Score:', r2)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
mse_test = mean_squared_error(Y_test_encoded, X_test_prediction)
print('Mean Squared Error:', mse_test)
```

```python
r2_test = r2_score(Y_test_encoded, X_test_prediction)
print('R-squared Score:', r2_test)
```
